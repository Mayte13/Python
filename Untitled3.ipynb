```python
df = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv",parse_dates = True, index_col=0)
```

*Mann-Kendal test*

La prueba no paramétrica de Mann-Kendall (*MK*) es ampliamente utilizada en el análisis de tendencias debido a su robustez y simplicidad.

```python
pip  install pymannkendall
```

```python
import pymannkendall as mk
```

```python
mk_O3 = mk.original_test(df['O3'], alpha=0.05)
```

```python
mk_O3
```

```python
a=mk_O3.slope
B=mk_O3.intercept
```

```python
def f5(x):
  return x*a+B
```

```python
x= range(0,3073)
fig, ax=plt.subplots(figsize=(10,5),dpi=200)
ax.plot(x,df['O3'],lw=1,color='pink');
ax.plot(x, [f5(i) for i in x], lw=4,color='purple')
```

```python
x= range(0,3073)
fig, ax=plt.subplots(figsize=(10,5),dpi=200)
ax.plot(x,df['PM10'],lw=1,color='orange');
ax.plot(x, [f5(i) for i in x], lw=4,color='red')
```
